```python
def get_speed(time_str):
    if time_str.count(':') == 2:
        h, m, s = time_str.split(':')
        total_secs = int(h) * 3600 + int(m) * 60 + int(s)
    elif time_str.count(':') == 1:
        m, s = time_str.split(':')
        total_secs = int(m) * 60 + int(s)
    else:
        total_secs = float(time_str)
    speed = round((3600 / total_secs),2)
    return speed
```
